```python
import numpy as np
```

using sklearn, the emptyfields in the column age are filled with the mean value of all ages

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer (missing_values = np.nan, strategy = "mean")
imputer = imputer.fit(df1[['Age']])
df1['Age'] = imputer.transform(df1[['Age']])
```

```python
df1.isnull().sum()
```

drop the column cabin which had the highest number of null data.

```python
df1=df1.drop(['Cabin'],axis=1)
```

```python
df1.head()
```

dropna drops all rows with atleast 1 null value

```python
df1=df1.dropna()
```

```python
df1.info()
```

we drop these columns as they do not help us infer to any decision in our model

```python
df1=df1.drop(['PassengerId','Name','Ticket','Embarked'],axis=1)
```
